Translate the column names of the DataFrame `df` from English to Brazilian Portuguese, then display the updated column names.

## Translate Column Names

### Subtask:
Create a dictionary mapping English column names to Brazilian Portuguese and then rename the columns of the DataFrame `df` using this dictionary.


**Reasoning**:
To rename the columns of the DataFrame as requested, I will first create a dictionary that maps the current English column names to their Brazilian Portuguese equivalents. Then I will use this dictionary with the `rename()` method to update the DataFrame's column names.



```python
new_columns = {
    'year': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_empregado': 'residencia',
    'taxa_remoto': 'remoto',
    'local_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}
df.rename(columns=new_columns, inplace=True)
print("DataFrame columns renamed successfully.")
print(df.columns)
```

```python
df["senioridade"]. value_counts()
```

```python
df["contrato"]. value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
print("Categorias da coluna 'senioridade' traduzidas com sucesso.")
print(df['senioridade'].value_counts())
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
print("Categorias da coluna 'contrato' traduzidas com sucesso.")
print(df['contrato'].value_counts())
```

```python
tamanho_empresa_traducao = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_traducao)
print("Categorias da coluna 'tamanho_empresa' traduzidas com sucesso.")
print(df['tamanho_empresa'].value_counts())
```

```python
remoto_traducao = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remoto_traducao)
print("Categorias da coluna 'remoto' traduzidas com sucesso.")
print(df['remoto'].value_counts())
```
