```python
import pandas as pd

# Recarrega o DataFrame para garantir que estamos trabalhando com os dados mais recentes
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")

# Dicionário de tradução das colunas (repetido para garantir a consistência neste bloco)
new_columns = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Renomeia as colunas do DataFrame
df = df.rename(columns=new_columns)

# Agora, execute o value_counts() na coluna 'contrato'
display(df["contrato"].value_counts())
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
print("Categorias da coluna 'senioridade' traduzidas com sucesso.")
print(df['senioridade'].value_counts())
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
print("Categorias da coluna 'contrato' traduzidas com sucesso.")
print(df['contrato'].value_counts())
```

```python
tamanho_empresa_traducao = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_traducao)
print("Categorias da coluna 'tamanho_empresa' traduzidas com sucesso.")
print(df['tamanho_empresa'].value_counts())
```

```python
remoto_traducao = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remoto_traducao)
print("Categorias da coluna 'remoto' traduzidas com sucesso.")
print(df['remoto'].value_counts())
```

```python
import numpy as np

#Criação de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000,  np.nan, 5000, np.nan, 100000]
})
#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios ['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''Calcula mediana e substitui os nulos pela mediana
'''

df_salarios ['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
import pandas as pd

# Carrega o DataFrame df
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")

# Cria df_limpo removendo linhas com valores nulos
df_limpo = df.dropna()

# Verifica a soma de valores nulos em df_limpo com a sintaxe correta
display(df_limpo.isnull().sum())
```

```python
import plotly.express as px
```

```python
# Calcula a média salarial por senioridade
media_salarial_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reindex(ordem).reset_index()

# Cria o gráfico de barras interativo com Plotly Express
fig = px.bar(
    media_salarial_senioridade,
    x='senioridade',
    y='usd',
    title='Média Salarial por Senioridade (USD)',
    labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual (USD)'},
    color='senioridade',
    color_discrete_sequence=px.colors.qualitative.Set2
)

# Exibe o gráfico
fig.show()
```
